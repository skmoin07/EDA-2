# ISOLATION FOREST (OUTLIERS) 

```python
from sklearn.ensemble import IsolationForest
```

```python
data
```

```python
#training model
clf=IsolationForest(random_state=10,contamination=0.1)
clf.fit(data)
```

```python
#prediction
y_preds_outliers=clf.predict(data)
```

```python
#-1 for outliers and 1 for inliers
y_preds_outliers
```
